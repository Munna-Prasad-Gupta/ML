```python
import numpy as np

def generate_classes_and_compute_distance(num_vectors, num_features, seed=0):
    """
    Generate random feature vectors for two classes, calculate centroids and spreads,
    and compute the distance between the centroids.

    Parameters:
    - num_vectors (int): Number of vectors for each class
    - num_features (int): Number of features for each vector
    - seed (int, optional): Random seed for reproducibility (default=0)

    Returns:
    - distance (float): Distance between the centroids of the two classes
    """
    np.random.seed(seed)

    class1_vectors = np.random.rand(num_vectors, num_features)
    class2_vectors = np.random.rand(num_vectors, num_features)

    centroid1 = np.mean(class1_vectors, axis=0)
    centroid2 = np.mean(class2_vectors, axis=0)

    spread1 = np.std(class1_vectors, axis=0)
    spread2 = np.std(class2_vectors, axis=0)

    distance = np.linalg.norm(centroid1 - centroid2)

    return distance

distance = generate_classes_and_compute_distance(100, 5)
print(distance)
```

```python
import numpy as np

def compute_minkowski_distances(vector1, vector2, max_r=10):
    """
    Compute Minkowski distances between two vectors for different values of r.

    Parameters:
    - vector1 (array): First vector
    - vector2 (array): Second vector
    - max_r (int, optional): Maximum value of r (default=10)

    Returns:
    - minkowski_distances (list): List of Minkowski distances for r=1 to max_r
    """
    minkowski_distances = []
    for r in range(1, max_r + 1):
        distance = np.linalg.norm(vector1 - vector2, ord=r)
        minkowski_distances.append(distance)

    return minkowski_distances

vector1 = class1_vectors[0]  # First vector of class 1
vector2 = class2_vectors[0]  # First vector of class 2

minkowski_distances = compute_minkowski_distances(vector1, vector2)
# print(minkowski_distances)

plt.plot(range(1, 11), minkowski_distances)
plt.xlabel('r')
plt.ylabel('Minkowski Distance')
plt.title('Minkowski Distance between Two Vectors')
plt.show()

```

```python
from sklearn.model_selection import train_test_split

X = np.vstack((class1_vectors, class2_vectors))  # Combine the feature vectors
y = np.hstack((np.zeros(100), np.ones(100)))  # Labels: 0 for class 1, 1 for class 2

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
accuracy = knn.score(X_test, y_test)
print(f'Accuracy: {accuracy:.2f}')
```

```python
predictions = knn.predict(X_test)
print('Predictions:', predictions)
```

```python
accuracies = []
for k in range(1, 12):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    accuracies.append(knn.score(X_test, y_test))

plt.plot(range(1, 12), accuracies)
plt.xlabel('k')
plt.ylabel('Accuracy')
plt.title('kNN Accuracy for Different k Values')
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, classification_report

y_pred = knn.predict(X_test)
conf_matrix = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:\n', conf_matrix)

report = classification_report(y_test, y_pred)
print('Classification Report:\n', report)
```
